```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = r"C:\Users\holai\OneDrive\Escritorio\Semestre 5\Modelos Computacionales I\Neurociencias-2025-1\datasets\cancer_regression\cancer_reg.csv"
df_cancer = pd.read_csv(csv_path, sep=",")
```

```python
# Histograma Método 1: Pandas
df_cancer.hist(column='medianage', bins=1000)
# El título se pone automático 
plt.xlabel('Edad Media')
plt.ylabel('Frecuencia')
```

```python
# Histograma Método 2: Matplotlib
plt.hist(df_cancer['medianage'], bins=1000)
plt.title('Distribución de la Edad Media')
plt.xlabel('Edad Media')
plt.ylabel('Frecuencia')
plt.show()
# Aquí no hay ningún label automático
```

```python
# Histograma Método 3: Seaborn
sns.histplot(df_cancer['medianage'], bins=1000)
plt.title('Distribución de la Edad Media')
plt.show()
# Los labels de los ejes se ponen en automático con seaborn 
```
